```python
from google_images_download import google_images_download 
```

```python
categories= ['automotive','beauty','sport','perfumes','shoes','military','woman','man','kids','digital_electronics']

def download_images(categories):
    response = google_images_download.googleimagesdownload()
    for category in categories:
        arguments = {"keywords":category,"limit":100,"format":"jpg","output_directory": "Images"} #, "print_urls":True}
        paths = response.download(arguments)
        
def json_images(categories):
    response = google_images_download.googleimagesdownload()
    for category in categories:
        #file = open(category+".txt", "w")
        arguments = {"keywords":category,"limit":100,"format":'jpg',# "print_urls":True,
                     "extract_metadata": True, "no_download": True, 
                     "chromedriver": "C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe"}
        paths = response.download(arguments)
        #file.close()
```
